```python
from sklearn.preprocessing import StandardScaler
```

```python
Y=data['Sold']
X=data.drop(['Sold'],axis=1)
```

```python
X.shape
```

```python
Y=np.array(Y)
Y.shape
```

```python
x_std=StandardScaler().fit_transform(X)
x_std.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(x_std,Y, test_size=0.2, random_state=10)
```

```python
clf=LogisticRegression()
```

```python
clf.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred=clf.predict(X_test)
```

```python
y_pred
```

```python
print(clf.score(X_train,Y_train))
clf.score(X_test,Y_test)
```

```python
accuracy_score(Y_test,y_pred)
```

```python
np.random.seed(42)
tf.random.set_seed(42)
```

```python
model=keras.models.Sequential([
    keras.layers.Dense(64, activation='linear',input_shape=[15]),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', 
              optimizer=keras.optimizers.SGD(lr=1e-1), 
              metrics=['accuracy'])
```

```python
model_history = model.fit(X_train, Y_train, epochs=100, validation_data=(X_test, Y_test))
```

```python
mae_test = model.evaluate(X_test, Y_test)
```

```python
import matplotlib.pyplot as plt
pd.DataFrame(model_history.history).plot(figsize=(8, 5))
plt.grid(True)
plt.gca().set_ylim(0, 1)

plt.show()
```
